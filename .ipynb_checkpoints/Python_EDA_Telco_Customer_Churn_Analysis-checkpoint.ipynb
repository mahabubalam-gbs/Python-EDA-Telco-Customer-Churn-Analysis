# Replacing blanks in 'TotalCharges' with zero(0) as tenure is zero(0) and no TotalCharges are recorded

```python
df['TotalCharges'] = df['TotalCharges'].replace(' ', '0')
df['TotalCharges'] = df['TotalCharges'].astype('float')
```

```python
def conv(value):
    if value == 1:
        return 'Yes'
    else:
        return 'No'

df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```
